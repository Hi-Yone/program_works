## ライブラリインポート

```python
import numpy as np
import matplotlib.pyplot as plt
```

## パラメータ設定

```python
D = 67  # 基準となる視距離[cm]
I = 6.5 # 眼間距離[cm]
disparity = [-0.7, -0.5, -0.3, -0.1, 0.0, 0.1, 0.3, 0.5, 0.7]   # 両眼視差
stim_radius = [1.0, 2.0, 4.0, 6.0, 8.0]                         # 刺激半径
```

## 視差と視距離の関係の計算

```python
theta = np.arctan(I/2 / D) * 180/np.pi  # 輻輳角
distance = [I/2 / np.tan((theta-disparity[ii]/2)*np.pi/180) for ii in range(len(disparity))]    # 視差を視距離に換算
distance = np.round(distance, 3)        # 3桁に丸める
print(distance)

# 視差と視距離の関係をプロット
plt.figure()
plt.plot(disparity, distance)
plt.xlabel('Disparity [deg]')
plt.ylabel('Visual distance [cm]')
plt.xticks(disparity)
plt.yticks(distance)
plt.grid()
```

## 大きさー奥行きのSSVEP計測実験の理論値を算出

```python
plt.figure()

for dist_ii in range(len(distance)):
    R_size = [np.arctan(stim_rr*2/distance[dist_ii]) * 180/np.pi for stim_rr in stim_radius]    # 刺激半径ごとに網膜像サイズを計算(理論値)
    
    plt.plot(R_size, stim_radius, 'o-', markersize=3, label=f"{disparity[dist_ii]} [deg]")
    
plt.ylabel('Object size (Stimulus radius) [deg]', fontsize=16)
plt.xlabel('Retinal size [deg]', fontsize=16)
plt.grid()
plt.legend(title='Disparity')
```
